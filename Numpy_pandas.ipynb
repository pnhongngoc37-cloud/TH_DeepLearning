```python
import numpy as np
def bai1():
    print("=== Bài 1 ===")
    arr = np.random.randint(0, 21, size=(3, 3))
    print("Mảng tạo được:\n", arr)
    # Kiểm tra tất cả phần tử khác 0
    result = np.all(arr != 0)
    print("Tất cả các phần tử khác 0?", result)
    return arr

if __name__ == "__main__":
    bai1()
```

```python
def bai3():
    print("\n=== Bài 3 ===")
    # Mảng 1 chiều
    a1 = np.array([1, 2, 3, 4])
    b1 = np.array([2, 2, 2, 5])
    print("Mảng 1D a1:", a1)
    print("Mảng 1D b1:", b1)
    print("a1 > b1 (greater):", np.greater(a1, b1))
    print("a1 >= b1 (greater_equal):", np.greater_equal(a1, b1))
    print("a1 < b1 (less):", np.less(a1, b1))
    print("a1 <= b1 (less_equal):", np.less_equal(a1, b1))

    # Mảng 2 chiều
    a2 = np.array([[1, 3], [5, 7]])
    b2 = np.array([[2, 2], [6, 6]])
    print("\nMảng 2D a2:\n", a2)
    print("Mảng 2D b2:\n", b2)
    print("a2 > b2 (greater):\n", np.greater(a2, b2))
    print("a2 >= b2 (greater_equal):\n", np.greater_equal(a2, b2))
    print("a2 < b2 (less):\n", np.less(a2, b2))
    print("a2 <= b2 (less_equal):\n", np.less_equal(a2, b2))
if __name__ == "__main__":
    bai3()
```

```python
import numpy as np
import pandas as pd

def bai20():
    print("\n=== Bài 20 ===")
    x = pd.Series(np.random.randint(1, 50, 20))
    y = pd.Series(np.random.randint(1, 50, 20))
    print("Series x:\n", x.values)
    print("Series y:\n", y.values)
    union = np.union1d(x, y)
    intersect = np.intersect1d(x, y)
    print("Union (hợp) của x và y:\n", union)
    print("Intersect (giao) của x và y:\n", intersect)

if __name__ == "__main__":
    bai20()
```

```python
import pandas as pd

def bai21():
    print("\n=== Bài 21 ===")
    x = pd.Series(range(10))
    y = pd.Series(['p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y'])
    print("Series x:\n", x)
    print("Series y:\n", y)
    # Theo chiều dọc (vertical) - dùng concat
    vertical = pd.concat([x, y], axis=0, ignore_index=True)
    print("Thêm theo chiều dọc (vertical):\n", vertical)
    # Theo chiều ngang (horizontal) - dùng concat
    horizontal = pd.concat([x, y], axis=1)
    print("Thêm theo chiều ngang (horizontal):\n", horizontal)

if __name__ == "__main__":
    bai21()
```

```python
import pandas as pd
import numpy as np

def bai29():
    print("\n=== Bài 29 ===")
    exam_data = {
        'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
        'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
        'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']
    }
    labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
    df = pd.DataFrame(exam_data, index=labels)
    result = df[(df['attempts'] > 2) & (df['score'].between(15, 20, inclusive='neither'))]
    print("Những dòng có attempt > 2 và score trong khoảng (15, 20):\n", result)

if __name__ == "__main__":
    bai29()
```

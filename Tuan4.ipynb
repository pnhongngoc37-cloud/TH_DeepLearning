```python
## =====================================================================
## THỰC HÀNH 4: CONVOLUTIONAL NEURAL NETWORK (CNN) - TẬP DỮ LIỆU MNIST
## =====================================================================

# 2.1. Khai báo thư viện phục vụ cho bài thực hành mạng CNN
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import load_model

# 2.2. Tải tập dữ liệu MNIST và phân chia dữ liệu train, test
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# 2.3. Hiển thị ngẫu nhiên các ảnh mẫu trong tập dữ liệu MNIST
plt.figure(figsize=(10, 5))
for i in range(10):
    plt.subplot(2, 5, i + 1)
    plt.imshow(X_train[i], cmap='gray')
    plt.title(f"Nhãn: {y_train[i]}")
    plt.axis('off')
plt.tight_layout()
plt.show()

# 2.4. Tiền xử lý dữ liệu ảnh cho mạng CNN
# Định dạng lại kích thước ảnh đầu vào (thêm kênh màu đơn sắc - channel=1)
X_train = X_train.reshape(X_train.shape[0], 28, 28, 1)
X_test = X_test.reshape(X_test.shape[0], 28, 28, 1)

# Chuyển đổi kiểu dữ liệu sang float32
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

# Chuẩn hóa giá trị pixel về đoạn [0, 1]
X_train /= 255
X_test /= 255

# Mã hóa One-hot cho nhãn mục tiêu (10 lớp từ 0-9)
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# 2.5. Khởi tạo kiến trúc mạng CNN thử nghiệm
model = Sequential()

# Tầng Convolutional thứ nhất + Tầng Max Pooling
model.add(Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(MaxPooling2D(pool_size=(2, 2)))

# Tầng Convolutional thứ hai + Tầng Max Pooling
model.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

# Làm phẳng đặc trưng thu được thành vector 1D
model.add(Flatten())

# Tầng Ẩn Fully Connected + Kỹ thuật Dropout chống quá khớp (Overfitting)
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))

# Tầng Đầu ra dự báo (Sử dụng hàm kích hoạt Softmax cho bài toán phân loại đa lớp)
model.add(Dense(10, activation='softmax'))

# Hiển thị cấu trúc tóm tắt của mạng CNN vừa xây dựng
model.summary()

# 2.6. Cấu hình các tham số tối ưu và biên dịch mô hình CNN
model.compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

# 2.7. Huấn luyện mô hình mạng CNN thử nghiệm
# Lưu thông tin quá trình học vào biến model_fit để vẽ đồ thị
model_fit = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=15,
    batch_size=128,
    verbose=1
)

# --- Vẽ đồ thị biểu diễn Độ chính xác (Accuracy) ---
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(model_fit.history['accuracy'])
plt.plot(model_fit.history['val_accuracy'])
plt.title('Độ chính xác của mô hình (Model Accuracy)')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')

# --- Vẽ đồ thị biểu diễn Hàm mất mát (Loss) ---
plt.subplot(1, 2, 2)
plt.plot(model_fit.history['loss'])
plt.plot(model_fit.history['val_loss'])
plt.title('Hàm mất mát của mô hình (Model Loss)')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')

plt.tight_layout()
plt.show()

# 2.8. Đánh giá mô hình CNN thử nghiệm trên tập test độc lập
score = model.evaluate(X_test, y_test, verbose=2)
print('\n--- KẾT QUẢ ĐÁNH GIÁ TRÊN TẬP TEST ---')
print('Test loss:', score[0])
print('Test accuracy:', score[1])

# 2.9. Dự báo nhãn cho ảnh sử dụng CNN
# Lấy ngẫu nhiên một ảnh từ tập test để kiểm tra khả năng dự đoán của mô hình
index = 99  # Bạn có thể thay đổi số này để check các ảnh khác nhau
test_image = X_test[index].reshape(1, 28, 28, 1)

# Thực hiện dự đoán xác suất cho từng lớp
prediction = model.predict(test_image)
# Lấy ra nhãn có xác suất cao nhất
predicted_label = np.argmax(prediction)
true_label = np.argmax(y_test[index])

# Hiển thị kết quả kiểm thử trực quan
plt.figure(figsize=(3, 3))
plt.imshow(X_test[index].reshape(28, 28), cmap='gray')
plt.title(f"Thực tế: {true_label} | Dự đoán: {predicted_label}")
plt.axis('off')
plt.show()

# 2.10. Lưu tham số mô hình CNN xuống ổ đĩa
# Định dạng file khuyến nghị hiện tại của Keras/Tensorflow là '.keras' (hoặc '.h5')
model.save('cnn_mnist_model.keras')
print("Đã lưu mô hình thành công vào file 'cnn_mnist_model.keras'")

# 2.11. Tải (Load) lại mô hình CNN đã lưu để tái sử dụng
loaded_model = load_model('cnn_mnist_model.keras')
print("Đã tải mô hình đã lưu lên hệ thống thành công!")

# Thử nghiệm nhanh mô hình vừa tải lại để chứng minh nó hoạt động bình thường
test_score = loaded_model.evaluate(X_test, y_test, verbose=0)
print(f"Độ chính xác của mô hình sau khi load lại: {test_score[1] * 100:.2f}%")
```

```python
## =====================================================================
## BÀI TẬP 1: XÂY DỰNG MẠNG CNN NHẬN DẠNG BỘ DỮ LIỆU CIFAR-10
## =====================================================================

# 1. Nạp các thư viện cần thiết
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPool2D, Flatten, Dense, Dropout, Input
from tensorflow.keras.utils import to_categorical

# Định nghĩa danh sách các nhãn tương ứng theo yêu cầu đề bài
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']
num_classes = 10

# 2. Đọc dữ liệu CIFAR-10 từ hệ thống TensorFlow
(X_train, y_train), (X_test, y_test) = cifar10.load_data()

print("Kích thước tập Train:", X_train.shape, y_train.shape)
print("Kích thước tập Test:", X_test.shape, y_test.shape)

# 3. Trực quan hóa một số ảnh mẫu đầu tiên
plt.figure(figsize=(12, 5))
for i in range(10):
    plt.subplot(2, 5, i + 1)
    plt.imshow(X_train[i])
    # y_train[i][0] để lấy giá trị nhãn phẳng thay vì mảng 2D
    plt.title(class_names[y_train[i][0]])
    plt.axis('off')
plt.tight_layout()
plt.show()

# 4. Tiền xử lý dữ liệu ảnh (Chuẩn hóa & Mã hóa One-hot)
# Ảnh CIFAR-10 có kích thước sẵn là (32, 32, 3) - Ảnh màu 3 kênh RGB
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0

# Chuyển giá trị nhãn thành dạng one-hot encoding
y_train_onehot = to_categorical(y_train, num_classes)
y_test_onehot = to_categorical(y_test, num_classes)

# 5. Xây dựng cấu trúc mô hình CNN
# Vì ảnh CIFAR-10 phức tạp hơn ảnh chữ số MNIST, cấu trúc mạng cần sâu hơn để đạt độ chính xác tốt
model = Sequential()
model.add(Input(shape=(32, 32, 3)))

# Block Convolution 1
model.add(Conv2D(32, kernel_size=(3, 3), padding='same', activation='relu'))
model.add(Conv2D(32, kernel_size=(3, 3), activation='relu'))
model.add(MaxPool2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

# Block Convolution 2
model.add(Conv2D(64, kernel_size=(3, 3), padding='same', activation='relu'))
model.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
model.add(MaxPool2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

# Làm phẳng dữ liệu và đưa vào các tầng liên kết đầy đủ (Fully Connected)
model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(num_classes, activation='softmax'))

# Hiển thị bảng tóm tắt kiến trúc mạng
model.summary()

# 6. Biên dịch mô hình
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# 7. Huấn luyện mô hình CNN (Chạy thử nghiệm 15 epochs)
model_fit = model.fit(
    X_train, y_train_onehot,
    batch_size=64,
    epochs=15,
    validation_split=0.1,
    verbose=1
)

# 8. Vẽ đồ thị Đánh giá mô hình (Độ chính xác và Hàm mất mát)
plt.figure(figsize=(12, 4))

# Đồ thị độ chính xác
plt.subplot(1, 2, 1)
plt.plot(model_fit.history['accuracy'], label='Train Accuracy')
plt.plot(model_fit.history['val_accuracy'], label='Val Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Đồ thị hàm mất mát
plt.subplot(1, 2, 2)
plt.plot(model_fit.history['loss'], label='Train Loss')
plt.plot(model_fit.history['val_loss'], label='Val Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()

# 9. Đánh giá mô hình trên tập kiểm thử (Test dataset)
score = model.evaluate(X_test, y_test_onehot, verbose=2)
print(f"\nĐộ chính xác trên tập kiểm thử (Test Accuracy): {score[1]*100:.2f}%")

# 10. Dự báo nhãn ngẫu nhiên thử nghiệm một bức ảnh từ tập Test
index = 100 # Chọn một tấm ảnh bất kỳ trong tập test
test_img = X_test[index].reshape(1, 32, 32, 3)

# Dự đoán
predict_probs = model.predict(test_img)
predicted_label_idx = np.argmax(predict_probs)
true_label_idx = y_test[index][0]

# Hiển thị trực quan kết quả dự báo
plt.figure(figsize=(3, 3))
plt.imshow(X_test[index])
plt.title(f"Thực tế: {class_names[true_label_idx]}\nĐoán: {class_names[predicted_label_idx]}")
plt.axis('off')
plt.show()

# 11. Lưu mô hình lại để chuẩn bị phục vụ triển khai Web Flask ở Câu 5
model.save('cifar10_cnn_model.h5')
print("Đã lưu mô hình CIFAR-10 thành công vào file 'cifar10_cnn_model.h5'")
```

```python
## =====================================================================
## BÀI TẬP 2: XÂY DỰNG MẠNG CNN NHẬN DẠNG BỘ DỮ LIỆU FASHION-MNIST
## =====================================================================

# 1. Khai báo các thư viện cần thiết
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Input
from tensorflow.keras.utils import to_categorical

# Định nghĩa 10 nhãn tương ứng với các loại trang phục
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
num_classes = 10

# 2. Tải tập dữ liệu Fashion-MNIST từ thư viện Keras
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

print("Kích thước tập huấn luyện (Train):", X_train.shape)
print("Kích thước tập kiểm thử (Test):", X_test.shape)

# 3. Trực quan hóa một số hình ảnh thời trang mẫu
plt.figure(figsize=(10, 5))
for i in range(10):
    plt.subplot(2, 5, i + 1)
    plt.imshow(X_train[i], cmap='gray')
    plt.title(class_names[y_train[i]])
    plt.axis('off')
plt.tight_layout()
plt.show()

# 4. Tiền xử lý dữ liệu (Reshape sang 4D và Chuẩn hóa pixel)
X_train = X_train.reshape(X_train.shape[0], 28, 28, 1).astype('float32') / 255.0
X_test = X_test.reshape(X_test.shape[0], 28, 28, 1).astype('float32') / 255.0

# Biến đổi nhãn sang dạng One-hot Encoding
y_train_onehot = to_categorical(y_train, num_classes)
y_test_onehot = to_categorical(y_test, num_classes)

# 5. Xây dựng cấu trúc mạng CNN
model = Sequential()
model.add(Input(shape=(28, 28, 1)))

# Lớp Chập 1 + Max Pooling + Chống quá khớp Dropout
model.add(Conv2D(32, kernel_size=(3, 3), activation='relu', padding='same'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

# Lớp Chập 2 + Max Pooling + Dropout
model.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

# Làm phẳng đặc trưng và kết nối với các lớp Dense ẩn
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))

# Lớp đầu ra phân loại 10 loại quần áo khác nhau
model.add(Dense(num_classes, activation='softmax'))

# In bảng cấu trúc mạng
model.summary()

# 6. Biên dịch mô hình
model.compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

# 7. Huấn luyện mô hình CNN (Chạy trong 12 epochs)
model_fit = model.fit(
    X_train, y_train_onehot,
    validation_data=(X_test, y_test_onehot),
    epochs=12,
    batch_size=128,
    verbose=1
)

# 8. Vẽ đồ thị biểu diễn Độ chính xác (Accuracy) và Hàm mất mát (Loss)
plt.figure(figsize=(12, 4))

# Đồ thị Accuracy
plt.subplot(1, 2, 1)
plt.plot(model_fit.history['accuracy'], label='Train Acc')
plt.plot(model_fit.history['val_accuracy'], label='Val Acc')
plt.title('Độ chính xác qua các Epoch')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Đồ thị Loss
plt.subplot(1, 2, 2)
plt.plot(model_fit.history['loss'], label='Train Loss')
plt.plot(model_fit.history['val_loss'], label='Val Loss')
plt.title('Hàm mất mát qua các Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()

# 9. Đánh giá độ chính xác tổng thể trên tập Test
score = model.evaluate(X_test, y_test_onehot, verbose=0)
print('\n--- KẾT QUẢ ĐÁNH GIÁ TRÊN TẬP TEST ---')
print(f'Mất mát (Test Loss): {score[0]:.4f}')
print(f'Độ chính xác (Test Accuracy): {score[1]*100:.2f}%')

# 10. Dự báo nhãn thử nghiệm ngẫu nhiên trên một sản phẩm
index = 250  # Lấy ngẫu nhiên phần tử thứ 251 trong tập test
test_image = X_test[index].reshape(1, 28, 28, 1)

# Dự đoán xác suất
prediction = model.predict(test_image)
predicted_label = np.argmax(prediction)
true_label = np.argmax(y_test_onehot[index])

# Hiển thị trực quan sản phẩm thời trang và kết quả dự đoán
plt.figure(figsize=(3, 3))
plt.imshow(X_test[index].reshape(28, 28), cmap='gray')
plt.title(f"Thực tế: {class_names[true_label]}\nDự đoán: {class_names[predicted_label]}")
plt.axis('off')
plt.show()

# 11. Lưu mô hình lại thành file cứng để phục vụ triển khai Web (Câu 5)
model.save('fashion_mnist_model.h5')
print("Đã lưu mô hình Fashion-MNIST thành công với tên 'fashion_mnist_model.h5'")
```

```python
## =====================================================================
## BÀI TẬP 3: XÂY DỰNG MẠNG CNN NHẬN DẠNG BỘ DỮ LIỆU FASHION-MNIST
## =====================================================================

# 1. Nạp các thư viện cần thiết
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Input
from tensorflow.keras.utils import to_categorical

# Định nghĩa danh sách các nhãn thời trang
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
num_classes = 10

# 2. Đọc dữ liệu Fashion-MNIST
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

# 3. Tiền xử lý dữ liệu
# Định dạng lại kích thước ảnh sang 4D (28x28x1) và chuẩn hóa pixel về [0, 1]
X_train = X_train.reshape(X_train.shape[0], 28, 28, 1).astype('float32') / 255.0
X_test = X_test.reshape(X_test.shape[0], 28, 28, 1).astype('float32') / 255.0

# Chuyển đổi nhãn sang One-hot encoding[cite: 1]
y_train_onehot = to_categorical(y_train, num_classes)
y_test_onehot = to_categorical(y_test, num_classes)

# 4. Xây dựng cấu trúc mô hình CNN[cite: 1]
model = Sequential([
    Input(shape=(28, 28, 1)),

    # Tầng Convolutional 1 + Max Pooling[cite: 1]
    Conv2D(32, kernel_size=(3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),

    # Tầng Convolutional 2 + Max Pooling[cite: 1]
    Conv2D(64, kernel_size=(3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),

    # Làm phẳng và các tầng Dense[cite: 1]
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5), # Kỹ thuật Dropout chống overfitting[cite: 1]
    Dense(num_classes, activation='softmax') # Softmax để phân lớp[cite: 1]
])

# 5. Biên dịch mô hình[cite: 1]
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# 6. Huấn luyện mô hình[cite: 1]
print("Bắt đầu huấn luyện...")
history = model.fit(X_train, y_train_onehot, epochs=10, batch_size=128, validation_split=0.1, verbose=1)

# 7. Đánh giá trên tập kiểm thử[cite: 1]
score = model.evaluate(X_test, y_test_onehot, verbose=0)
print(f'\nĐộ chính xác trên tập Test: {score[1]*100:.2f}%')

# 8. Trực quan hóa kết quả dự báo[cite: 1]
plt.figure(figsize=(10, 5))
for i in range(5):
    plt.subplot(1, 5, i + 1)
    img = X_test[i].reshape(28, 28)
    plt.imshow(img, cmap='gray')

    # Dự báo nhãn[cite: 1]
    pred = model.predict(X_test[i:i+1])
    pred_label = class_names[np.argmax(pred)]
    true_label = class_names[y_test[i]]

    plt.title(f"P: {pred_label}\nT: {true_label}", fontsize=9)
    plt.axis('off')
plt.tight_layout()
plt.show()

# 9. Lưu mô hình[cite: 1]
model.save('fashion_mnist_cnn.h5')
```

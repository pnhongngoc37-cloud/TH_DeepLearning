```python
import tensorflow as tf
import numpy as np
```

```python
# Bai12
# Tạo graph
graph = tf.Graph()
with graph.as_default():
    x = tf.random.normal([3,2,2], name='x')
    y = tf.random.normal([3,2,2], name='y')

    add_op = tf.add(x, y, name='add')
    sub_op = tf.subtract(x, y, name='sub')
    mul_op = tf.multiply(x, y, name='mul')
    div_op = tf.divide(x, y, name='div')

# Chạy graph
with tf.compat.v1.Session(graph=graph) as sess:
    sess.run(tf.compat.v1.global_variables_initializer())
    x_val, y_val, add_val, sub_val, mul_val, div_val = sess.run([x, y, add_op, sub_op, mul_op, div_op])
    print("x:\n", x_val)
    print("add:\n", add_val)
```

```python
# Bai13
graph = tf.Graph()
with graph.as_default():
    x = tf.compat.v1.placeholder(tf.float32, [3,2,2], name='x')
    y = tf.compat.v1.placeholder(tf.float32, [3,2,2], name='y')
    sum_op = tf.add(x, y, name='sum')

with tf.compat.v1.Session(graph=graph) as sess:
    x_val = np.random.randn(3,2,2)
    y_val = np.random.randn(3,2,2)
    result = sess.run(sum_op, feed_dict={x: x_val, y: y_val})
    print(result)
```
